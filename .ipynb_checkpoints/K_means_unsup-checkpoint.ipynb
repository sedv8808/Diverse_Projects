```python
###### BEGIN SOLUTION

n_sim = 100
n_clusters = 4

# Let's use the best inertia in a 100 runs of KMeans++ as the optimal inertia possible;
best_inertia = cluster.KMeans(n_clusters=4, n_init=100).fit(x).inertia_

# Let's create a function so we can reuse the code;
def get_info(n_cluster, n_sim, x, n_init, init):
    """
    This function will print the info requested by the exercise for the configuration provided in the parameters.\n 
    
    Parameters:
    ----------
    n_sim: the number of runs of K-Means (the exercise asked for 100);
    x: the data;
    n_init: the n_init value for each run of the algorithm (for the K-Means in this exercise, it will be [1,4,20])
    init: the ìnit parameter for each run of the algorithm - 'random' for K-Means and 'k-means++' for K-Means++
    """
    
    t = "++" if (init == "k-means++") else ""
    print(f"\n\n########################\nK-Means{t}(n_init={n_init})\n########################\n")
    
    #Let's start by timing
    start = time.time() 
    kmeans = list()
    for i in range(0,n_sim):
        kmeans.append(cluster.KMeans(n_clusters=n_clusters, init=init, n_init=n_init)) #It must use n_init=1
        kmeans[i].fit(x)
    end = time.time()
    print(f"It took approx. {np.round(end-start,4)} seconds to run K-Means{t}(n_init={n_init}) {n_sim} times.")

    # Now, let's get the summary of the inertia of the kmeans
    kmeans_inertia = pd.Series([x.inertia_ for x in kmeans])
    print(f"\nThe summary of inertia across {n_sim} executions of K-Means{t}(n_init={n_init}):\n")
    print(kmeans_inertia.describe())

    # Now, let's calculate the proportion that K-Means found the minimun
    kmeans_min_inertia = kmeans_inertia.min()
    count = 0
    for i in range(0, n_sim):
        if np.abs(kmeans_inertia[i] - best_inertia) < 1:
            count += 1
    prob = count/n_sim
    print(f"\nThe K-Means{t}(n_init={n_init}) found the optimal cluster in {100*prob}% of the times.")
    

# To get a better estimate of the probability of K-Means finding the optimal cluster, you can 
# increase the n_sim value. E.g., get_info(n_clusters, 15000, x, 1, 'random')
for i in [1,4,20]:
    get_info(n_clusters, n_sim, x, i, 'random')
    
get_info(n_clusters, n_sim, x, 1, 'k-means++')
```

##### Item C:

Since, the rate of success of K-Means(n\_init=$4$) is less the K-Means++(n\_init=1), it is better to go with the latter one. On the other hand, K-Means with `n_init=20` always find the optimal clustering, but takes more time to run, which might or might not be acceptable to you.  

##### item D:
Let $X$ be the probability that the K-Means finds the optimal solution in one run. So $X\sim Bernoulli(p)$. From Item 1, $p\approx 43\%$.

So, it follows that, the probability that the K-Means finds the optimal solution in 4 runs, is given by
$P\left(\sum_{i=1}^4 X_i > 0\right) = 1 - P\left(\sum_{i=1}^4 X_i = 0\right) = 1 - 0.10556 = 0.89444$, which is very close to the estimated 87%. [Note that $\sum_{i=1}^4 X_i\sim Binomial(4,0.43)$]

Last, the probability that the K-Means finds the optimal solution in 20 runs, is given by
$P\left(\sum_{i=1}^{20} X_i > 0\right) = 1 - P\left(\sum_{i=1}^{20} X_i = 0\right) = 1 - 1.310681e-05 = 0.9999869$, which is very close to the estimated 100%.

### Exercise 3 - Solution

```python
###### BEGIN SOLUTION
plt.figure(figsize=(22,10))
kmeans = list()
ss = list()
for k in range(0,6):
    kmeans.append(cluster.KMeans(n_clusters=k+2))
    kmeans[k].fit(x)
    ss.append(metrics.silhouette_score(x, kmeans[k].labels_))

inertia = [x.inertia_ for x in kmeans]
plt.plot(range(2,8),inertia)
plt.xlabel("k", fontdict={"fontsize":22})
plt.ylabel("Inertia", fontdict={"fontsize":22});
###### END SOLUTION
```

The Elbow plot presents a steep decrease of inertia from $k=2$ to $k=3$ which justify increasing the number of cluster. There's also a fair decrease from $k=3$ to $k=4$ although not that steep, which makes it questinable if another cluster should be considered or not. The decrease from $k=4$ to $k=5$ is very minor, suggesting that the number of cluster shouldn't be increased beyond 4.

```python
plt.figure(figsize=(22,10))
plt.plot(range(2,8),ss)
plt.xlabel("k", fontdict={"fontsize":22})
plt.ylabel("Silhouette Score", fontdict={"fontsize":22})
plt.axvline(x=np.argmax(ss)+2, linestyle='-.', c='black');
```

Since the highest Silhouette Score is obtained for $k=4$, this is the suggested number of cluster by the Silhouette method.

Discovering object groups is one motivation for clustering. Another motivation is _vector quantization_, where we find a prototype point for each cluster and replace points in the cluster by their prototype. If our inputs are images, vector quantization gives us a rudimentary image compression algorithm.

We will implement image quantization by filling in the `quantize` and `dequantize` functions below. The `quantize` function should take in an image and, using the pixels as examples and the 3 colour channels as features, run $k$-means clustering on the data with $2^b$ clusters for some hyperparameter $b$. The code should store the cluster means and return the cluster assignments. The `dequantize` function should return a version of the image (the same size as the original) where each pixel's orignal colour is replaced with the nearest prototype colour.

To understand why this is compression, consider the original image space. Say the image can take on the values $0,1,\ldots,254,255$ in each colour channel. Since $2^8=256$ this means we need 8 bits to represent each colour channel, for a total of 24 bits per pixel. Using our method, we are restricting each pixel to only take on one of $2^b$ colour values. In other words, we are compressing each pixel from a 24-bit colour representation to a $b$-bit colour representation by picking the $2^b$ prototype colours that are "most representative" given the content of the image. 

Your tasks:

1. complete the `quantize` and `dequantize` functions below.
2. run the code on an image of your choosing and display the results for a few different values of $b$.

(Note 1: if you actually try saving this as a file, you won't see the file size being what you expected, because Python won't know to allocate exactly $b$ bits per element of `quantized_img`, and will instead probably store them as 32-bit integers if you don't specify otherwise. But if one wanted to work harder, it would be theoretically possible to store the elements with $b$ bits per pixel. Also, all this is before any additional lossless compression).

(Note 2: this is not how image compression systems like JPEG actually work. It uses something similar to the Fourier transform, followed by a (simpler) quantization, followed by lossless compression.)

```python
import numpy as np
import os
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
```

```python
img = imread(os.path.join("data", "mandrill.jpg"))
imshow(img)
plt.title("original image")
plt.show()
```

```python
def quantize(img, b):
    """
    Quantizes an image into 2^b clusters

    Parameters
    ----------
    img : a (H,W,3) numpy array
    b   : an integer
    
    Returns
    -------
    quantized_img : a (H,W) numpy array containing cluster indices
    colours       : a (2^b, 3) numpy array, each row is a colour
    """

    H, W, _ = img.shape
    model = KMeans(n_clusters=2**b)

    ##### # TODO: fill in code here
    model.fit(img.reshape(H * W, 3));
    quantized_img = model.labels_.reshape(H,W).copy()
    #####
    
    return quantized_img, model.cluster_centers_.astype('uint8')

def dequantize(quantized_img, colours):
    H, W = quantized_img.shape
    img = np.zeros((H,W,3), dtype='uint8')

    ##### # TODO: fill in the values of `img` here
    img = np.array(colours[quantized_img,:], dtype='uint8')
    #####
    
    return img
```

```python
b = 3
compressed, colours = quantize(img, b = b)
recon = dequantize(compressed, colours)
fig, ax = plt.subplots(1,2, figsize=(10,10))
ax[0].imshow(img)
ax[0].set_title("original image")
ax[1].imshow(recon) 
ax[1].set_title(f"reconstructed image (b = {b})");
```
